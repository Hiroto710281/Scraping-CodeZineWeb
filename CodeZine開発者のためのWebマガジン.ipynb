```python
from datetime import datetime
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://codezine.jp/article/tag/223')
root_url = 'https://codezine.jp/'
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
articles = []
```

```python
#CSS <ul class="catList"><li> 
```

```python
lis = soup.select('ul.catList > li')
for li in lis:
    #get 
    day = li.find('div', class_='day').text.strip()
    #dattetime
    published = datetime.strptime(day, '%Y/%m/%d')
    h2_tag = li.find('h2') #h2tag
    title = h2_tag.text #title
    url = h2_tag.a['href'] #URL
    
    
    
    tag_list = li.select('ul.tag > li') #tag > li
    tags = [tag.text.strip() for tag in tag_list]#make list for tag
    
    article = {
        'published': published,
        'title': title,
        'url': root_url + url,
        'tags': tags
    }
    articles.append(article)
```
